```python
import time
from selenium import webdriver
from PIL import Image as pil
```

```python
# screenshot 
driver = webdriver.Chrome()
driver.get("http://www.youtube.com/watch?v=tXOJIBPwSyo&t=200s")
time.sleep(1)
driver.save_screenshot('screenshot.png')
```

```python
# Save Screenshot
pil_im = pil.open('screenshot.png')
print(pil_im.size)
pil_im
```

```python
# change image size
pil_im.thumbnail((512, 512))
pil_im.save("screenshot_thumbnail.png")
pil.open('screenshot_thumbnail.png')
```

```python
#   
element = driver.find_element_by_css_selector('#movie_player')
location = element.location
size = element.size
location, size
```

```python
pil_im = pil.open("screenshot.png")
left = location['x'] * 2
top = location['y'] * 2
right = location['x'] + size['width'] * 2
bottom = location['y'] + size['height'] * 2

area = (left, top, right, bottom)
area
```

```python
# .crop : cut image (4D tuple)
pil_im = pil_im.crop(area)
pil_im.save("screenshot_element.png")
pil.open('screenshot_element.png')
```
